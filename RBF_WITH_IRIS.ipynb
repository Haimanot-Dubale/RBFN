```python
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import numpy as np

datasets=load_iris()
```

```python
x_index = 0
y_index = 1

# this formatter will label the colorbar with the correct target names
formatter = plt.FuncFormatter(lambda i, *args: datasets.target_names[int(i)])

plt.figure(figsize=(5, 4))
plt.scatter(datasets.data[:, x_index], datasets.data[:, y_index], c=datasets.target)
plt.colorbar(ticks=[0, 1, 2], format=formatter)
plt.xlabel(datasets.feature_names[x_index])
plt.ylabel(datasets.feature_names[y_index])

plt.tight_layout()
plt.show()
```
